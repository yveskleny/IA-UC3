**EXERCÍCIO 01**

Crie uma função que receba dois números e retorne a multiplicação:

```python
def multiplica(a, b):
  """
  função que recebe dois números e retorna a multiplicação
  Parâmetros:
    a (int): primeiro número
    b (int): segundo número
  """
  return a*b

a = 5
b = 4
resultado = multiplica(a, b)

print(resultado)
```

**EXERCÍCIO 02:**

Crie uma função que calcule a média entre dois números

```python
def media(a, b):
  """
  Função que calcula a média entre dois números
  Parâmetros:
    a (int): primeiro número
    b (int): segundo número
  """
  return (a + b)/2

a = float(input("Digite o valor 1: "))
b = float(input("Digite o valor 2: "))

resultado = media(a, b)

print(f"A media é {resultado}")
```

**EXERCÍCIO 04:**

Crie uma função que retorna se um número é par ou impar

```python
def verifica_paridade(numero):
  """
  Função que retorna a paridade de um número:
  Parâmetro:
    numero (int)
  """
  if numero % 2 == 0:
    print(f"O numero {numero} é Par!")
  else:
    print(f"O numero {numero} é Impar!")


for i in range(10):
  verifica_paridade(i)
```

**EXERCÍCIO 05:**

 Crie uma função soma_imposto(taxa_imposto, custo) que altere o valor do custo incluindo o imposto (ex: taxa de 10% sobre 100 retorna 110)

```python
def soma_impostos(taxa_imposto, custo):
  """
  Função que adiciona uma taxa de imposto sobre um determinado custo
  Parâmetros:
    taxa_imposto: valor da taxa de importo em porcentagem
    custo: valor total do custo
  """
  return (custo + custo * (taxa_imposto / 100))


print(soma_impostos(15, 80))
```

EXERCÍCIO 06:
Crie uma função que tenha como parâmetro 4 notas de alunos e retorne se ele foi aprovado ou não (considerando que a média de aprovação seja 6)

```python

def media_alunos(nota1bi, nota2bi, nota3bi, nota4bi):
  soma = nota1bi + nota2bi + nota3bi + nota4bi
  media = soma/4

  if media >= 6:
    print(f"Media = {media}, Aprovado")
  else:
    print(f"Media = {media}, Reprovado")

from random import randint

media_alunos(randint(0, 10), randint(0, 10), randint(0, 10), randint(0, 10))
```
