## 1. Função Linear

### Forma geral

$$
f(x) = ax + b
$$

### Ideia principal
A função linear representa uma relação de crescimento ou decrescimento constante.

- `a` = taxa de variação
- `b` = valor inicial

### Exemplo
$$

$$

Se `x = 4`:

$$
f(4) = 2 \cdot 4 + 3 = 11
$$

## 5. Probabilidade Básica

### Fórmula

$$
P(A) = \frac{\text{casos favoráveis}}{\text{casos possíveis}}
$$

### Ideias principais
- probabilidade mede chance;
- eventos independentes: multiplicamos probabilidades;
- “e” é diferente de “ou”.

### Exemplos
- probabilidade de cara em uma moeda: \(1/2\)
- probabilidade de cara duas vezes seguidas: \((1/2) \cdot (1/2) = 1/4\)

### Onde isso aparece em IA?
- classificação;
- decisão baseada em probabilidades;
- Naive Bayes;
- previsão com incerteza.

```python
# Exemplo simples de decisão da IA com probabilidades

classes = ["A", "B", "C"]
probabilidades = [0.2, 0.6, 0.2]

maior_prob = probabilidades[0]
indice_maior = 0

for i in range(len(probabilidades)):
    if probabilidades[i] > maior_prob:
        maior_prob = probabilidades[i]
        indice_maior = i

decisao = classes[indice_maior]

print("Probabilidades:", probabilidades)
print("Decisão da IA:", decisao)
```

### Exercício 6
Considere a função:

g(x) = -x^2 + 4x + 1

Responda:

### Exercício 8
Crie uma função polinomial simples que represente uma situação em que um valor cresce no início e depois diminui.

```python
def lançamento_objeto(altura_inicial, velocidade_inicial, aceleracao_gravidade, tempo):
    resultado = (altura_inicial + velocidade_inicial*tempo - 0.5*aceleracao_gravidade*tempo**2)
    return resultado if resultado >= 0 else 0

tempos = np.linspace(0, 10, 100)
y = [lançamento_objeto(altura_inicial=10, velocidade_inicial=15, aceleracao_gravidade=10, tempo=t) for t in tempos]

plt.figure(figsize=(8, 5))
plt.plot(tempos, y)
idx = y.index(max(y))
plt.plot(tempos[idx] ,max(y), 'ro')
plt.text(tempos[idx] ,max(y), f'Altura máxima = {max(y):.2f} m²', ha="left")
plt.grid()
plt.show()
```

### Exercício 18
Qual a probabilidade de sair cara duas vezes seguidas ao lançar uma moeda duas vezes?

```python
# Probabilidade Cara = p
# Probabilidade Coroa = 1 - p

# Eventos provaveis:
#   Cara e Cara -> p * p
#   Cara e Coroa -> p * (1 - p)
#   Coroa e Cara -> (1 - p) * p
#   Coroa e Coroa -> (1 - p) * (1 - p)

# Assumindo que se trata de o lançamento de uma moeda honesta p = 0.5

p = 0.5
print(f"Probabilidade de Cara e Cara é {p * p}")
print(f"Probabilidade de Cara e Coroa é {p * (1 - p)}")
print(f"Probabilidade de Coroa e Cara é {(1 - p) * p}")
print(f"Probabilidade de Coroa e Coroa é {(1 - p) * (1 - p)}")
```
