# Revisão: Blocos, Condições e laços em python

## 1. Blocos de código em Python

Em Python, blocos são definidos pela indentação.

```python
if True:
    print("Este comando está dentro do bloco")
    print("Este também está dentro")

print("Este está fora do bloco")
```

Observe que os comandos dentro do if estão indentados.

Normalmente usamos 4 espaços.

**Exercício**

```python
x = 10

if x > 5:
    print("x é maior que 5")

print("Fim do programa")
```

## 2. Estruturas Condicionais

Permitem que o programa tome decisões.

Estrutura básica:

```python
if condição:
    código
```

Exemplo

```python
idade = 20

if idade >= 18:
    print("Maior de idade")
```

Usando else

```python
idade = 16

if idade >= 18:
    print("Maior de idade")
else:
    print("Menor de idade")
```

Usando elif

```python
nota = 7

if nota >= 7:
    print("Aprovado")
elif nota >= 5:
    print("Recuperação")
else:
    print("Reprovado")
```

**Exercício**

*   Positivo

*   zero

```python
numero = int(input("Digite um número: "))

if numero > 0:
  print(f"O numero {numero} é positivo!")
elif numero < 0:
  print(f"O numero {numero} é negativo!")
else:
    print(f"Zero!")
```

## 4. Laço while

O while executa enquanto uma condição for verdadeira.

```python
contador = 0

while contador < 5:
    print(contador)
    contador += 1
```

Exemplo

```python
numero = 1

while numero <= 10:
    print(numero)
    numero += 1
```

**Exercício**

Crie um while que mostre números de 10 até 1.

```python
# escreva aqui
contador = 10

while contador > 0:
  print(contador)
  contador -= 1
```

## 5. Controle de fluxo

**Break**

Interrompe o loop.

```python
for i in range(10):
    if i == 5:
        break
    print(i)
```

**Continue**

Pula uma iteração.

```python
for i in range(10):
    if i == 5:
        continue
    print(i)
```

## 7. Exemplo prático

Somar números de uma lista.

```python
numeros = [10, 20, 30, 40]

soma = 0

for n in numeros:
    soma += n

print("Soma:", soma)
```

**Exercício**

Calcule a média de uma lista de números.
Essa lista pode ter qualquer tamanho e qualquer número. Seu código tem que identificar o tamanho da lista automaticamente e calcular a média dos valores presentes.

```python
numeros = [1, 27, 60, 44, 13, 48]

soma = 0

for num in numeros:
    soma += num

print(f"media = {soma / len(numeros):.2f}")
```

# **8. Desafio Final**

Crie um programa que:

* peça 5 números ao usuário

* guarde em uma lista

* mostre:

  --  maior número

  -- menor número

  -- média

Estrutura sugerida:

while

lista

for

```python
lista_numeros = []

for i in range(5):
    num = float(input(f"Digite o {i + 1}º numero: "))
    lista_numeros.append(num)
print(f"O menor numero é {min(lista_numeros)}")
print(f"O maior numero é {max(lista_numeros)}")
print(f"A média dos números é {(sum(lista_numeros)/len(lista_numeros)):.2f}")
```

```python
lista = []

qtd_de_numeros = 5

while qtd_de_numeros > 0:
    lista.append(float(input("Digite um numero: ")))
    qtd_de_numeros -= 1

print(max(lista))
print(min(lista))
print(sum(lista)/len(lista))
```
