```python
import numpy as np

nota1 = np.array(df['nota1'])

media = nota1.mean()
maior_valor = nota1.max()
menor_valor = nota1.min()
soma_notas = nota1.sum()

print("Media:", media)
print("Maior nota:", maior_valor)
print("Menor nota:", menor_valor)
print("Soma das notas:", soma_notas)
```
