## 1. Revisão rápida: o que é uma função exponencial?

Uma função exponencial é uma função em que a variável aparece no **expoente**.

Forma geral:

\[
f(x) = a . b^x
\]

onde:

- `a` é o valor inicial;
- `b` é a base de crescimento ou decrescimento;
- `x` é a variável de entrada.

### Ideia principal

Enquanto em uma função linear o crescimento acontece por **soma constante**,  
na função exponencial o crescimento acontece por **multiplicação constante**.

---

## Comparando rapidamente

### Linear
\[
f(x) = 2x + 1
\]

Cresce somando sempre a mesma quantidade.

### Exponencial
\[
f(x) = 2^x
\]

Cresce multiplicando o valor cada vez mais rapidamente.

## 2. Exemplo do cotidiano

Imagine um conteúdo publicado em uma rede social.

No primeiro momento, poucas pessoas veem.  
Depois, algumas compartilham.  
Depois, essas pessoas compartilham com outras.

Em certos casos, o alcance pode crescer muito rápido.

Um modelo simplificado pode ser:

\[
A(x) = 100 * 2^x
\]

onde:

- `x` = número de etapas de compartilhamento;
- `A(x)` = quantidade estimada de pessoas alcançadas.

```python
# Exemplo 1: alcance de um conteúdo ao longo das etapas

def alcance(x):
    return 100 * (2 ** x)

etapas = [0, 1, 2, 3, 4, 5]
[(x, alcance(x)) for x in etapas]
```

## Resolução comentada do exemplo

Calculando:

- `A(0) = 100`
- `A(1) = 200`
- `A(2) = 400`
- `A(3) = 800`
- `A(4) = 1600`
- `A(5) = 3200`

Perceba que o crescimento não acontece somando sempre 100.  
Na verdade, o valor vai sendo **multiplicado por 2** a cada etapa.

## 3. Visualizando o crescimento exponencial

Vamos observar o gráfico do exemplo do alcance.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 6, 200)
y = 100 * (2 ** x)

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title("Crescimento exponencial do alcance")
plt.xlabel("Etapas")
plt.ylabel("Pessoas alcançadas")
plt.grid()
plt.show()
```

### O que observar no gráfico?

1. O crescimento é lento no início ou muito rápido desde o começo?
2. A curva parece reta ou curva?
3. O crescimento aumenta cada vez mais ou permanece constante?

## 4. Crescimento linear x crescimento exponencial

Essa comparação é muito importante para entender a diferença.

```python
# Comparando crescimento linear e exponencial

x = np.arange(0, 8)
linear = 100 + 100*x
exponencial = 100 * (2 ** x)

plt.figure(figsize=(8, 5))
plt.plot(x, linear, marker="o", label="Linear: 100 + 100x")
plt.plot(x, exponencial, marker="o", label="Exponencial: 100·2^x")
plt.title("Comparação entre crescimento linear e exponencial")
plt.xlabel("x")
plt.ylabel("Valor")
plt.legend()
plt.grid()
plt.show()
```

## Interpretação da comparação

- Na função linear, o crescimento acontece por **adição constante**.
- Na exponencial, o crescimento acontece por **multiplicação constante**.
- Por isso, a exponencial pode crescer muito mais rápido.

Essa diferença é importante em tecnologia, ciência de dados e IA.

## 5. Decrescimento exponencial

Nem toda função exponencial cresce.  
Ela também pode representar queda rápida.

Exemplo:

\[
d(x) = 1000 * (0.5)^x
\]

onde:

- `x` = tempo
- `d(x)` = quantidade restante

Nesse caso, a cada etapa, fica apenas metade do valor anterior.

```python
# Exemplo 2: decrescimento exponencial

def decrescimento(x):
    return 1000 * (0.5 ** x)

tempos = [0, 1, 2, 3, 4, 5]
[(x, decrescimento(x)) for x in tempos]
```

```python
# Gráfico do decrescimento exponencial

x = np.linspace(0, 6, 200)
y = 1000 * (0.5 ** x)

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title("Decrescimento exponencial")
plt.xlabel("Tempo")
plt.ylabel("Quantidade")
plt.grid()
plt.show()
```

## 6. Onde isso aparece em IA?

Aqui está a conexão principal com Inteligência Artificial.

### 1. Funções de ativação
Em redes neurais, algumas funções importantes usam a ideia de exponencial.

Exemplo: a função **sigmoid** usa exponencial para transformar valores em algo entre 0 e 1.

\[
\sigma(x) = \frac{1}{1 + e^{-x}}
\]

Você não precisa aprofundar agora.  


### 2. Probabilidades em classificação
Em vários modelos de IA, precisamos transformar saídas numéricas em probabilidades.  
Ideias exponenciais aparecem nesse processo.

### 3. Crescimento de dados e custo computacional
Na prática, alguns problemas podem crescer muito rápido conforme aumentam os dados, o número de combinações ou as possibilidades.

### 4. Aprendizado e ajuste de modelos
Em alguns contextos, o erro pode cair rapidamente no início e depois mudar de ritmo.  
Além disso, conceitos exponenciais aparecem em fórmulas e algoritmos de otimização.

## 7. Mostrando um exemplo ligado à IA

Vamos mostrar uma versão simplificada da função sigmoid, muito usada em IA.

```python
# Exemplo 3: função sigmoid

x = np.linspace(-10, 10, 400)
sigmoid = 1 / (1 + np.exp(-x))

plt.figure(figsize=(8, 5))
plt.plot(x, sigmoid)
plt.title("Função Sigmoid")
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.grid()
plt.show()
```

## Como explicar isso para os alunos



> Essa curva aparece em IA para transformar valores em algo parecido com probabilidade.

Exemplo de leitura:

- se a saída estiver perto de `0`, a chance é baixa;
- se estiver perto de `1`, a chance é alta.

Não se preocupem, essa função será aprofundada em Redes Neurais

## 8. Atividade prática com Python

Agora vamos trabalhar com uma situação aplicada.

Imagine que a quantidade de dados processados por um sistema dobre a cada ciclo.

Modelo:

\[
q(x) = 50 * 2^x
\]

onde:

- `x` = número de ciclos
- `q(x)` = quantidade de dados

```python
# Exemplo 4: quantidade de dados ao longo dos ciclos

def quantidade_dados(x):
    return 50 * (2 ** x)

for x in range(6):
    print(f"Ciclo {x}: {quantidade_dados(x)} unidades de dados")
```

```python
# Gráfico da quantidade de dados

x = np.linspace(0, 6, 200)
y = 50 * (2 ** x)

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title("Crescimento da quantidade de dados")
plt.xlabel("Ciclos")
plt.ylabel("Dados")
plt.grid()
plt.show()
```
