# Princípio Fundamental da Contagem (PFC)

##  O que é?

O Princípio Fundamental da Contagem diz:

> Se um processo ocorre em etapas independentes, multiplicamos as possibilidades de cada etapa.

---

## Intuição

Se você tem escolhas em sequência, você **multiplica**.

---



##  Exemplo 1 — Roupas

- 3 camisetas  
- 2 calças  

Quantos looks diferentes?

$$
3 \times 2 = 6
$$

---



##  Exemplo 2 — Moeda + Dado

- Moeda → 2 opções (cara, coroa)  
- Dado → 6 opções  

Total de possibilidades:

$$
2 \times 6 = 12
$$

---



##  Conexão com Probabilidade

A probabilidade é dada por:

$$
P(A) = \frac{\text{casos favoráveis}}{\text{casos possíveis}}
$$

👉 O PFC nos ajuda a encontrar o **total de casos possíveis**.

---



## 📌 Exemplo aplicado

Qual a probabilidade de tirar **cara e depois o número 6**?

- Total de possibilidades: 12  
- Casos favoráveis: 1  

$$
P = \frac{1}{12}
$$


## 5. Eventos dependentes (intuição)

Se retirar carta sem reposição → muda a probabilidade

Isso é MUITO importante para IA (dados mudam!)


EXEMPLO 1: CARTAS

“Se eu tiro uma carta e NÃO devolvo… o baralho muda.”

```python
cartas = ["❤️", "❤️", "♠", "♠"]
```

Qual a probabilidade de tirar duas cartas ❤️ seguidas sem reposição?

1ª carta:

* 2/4 = 1/2

2ª carta:

agora só sobram 3 cartas
apenas 1 é ❤️

* 1/3

P = 1/2 * 1/3 = 1/6

```python
import random

cartas = ["❤️", "❤️", "♠", "♠"]

sucesso = 0
n = 10000

for _ in range(n):
    baralho = cartas.copy()
    c1 = random.choice(baralho)
    baralho.remove(c1)

    c2 = random.choice(baralho)

    if c1 == "❤️" and c2 == "❤️":
        sucesso += 1

sucesso / n
```


## 6. Frequência vs probabilidade

Vamos simular:


```python

import random

n=10000
caras=0

for _ in range(n):
    if random.choice(["cara","coroa"])=="cara":
        caras+=1

caras/n
```
