## 1. Revisão rápida: o que é logaritmo?

O logaritmo responde à pergunta:

> **“A que potência precisamos elevar uma base para obter determinado valor?”**

Exemplo:


$$\log_2(8) = 3$$


Isso significa que:

$$
2^3 = 8
$$

---

## Ideia central

Logaritmo e exponencial são operações inversas.

Se:

$$
2^3 = 8
$$

então:

$$
\log_2(8) = 3
$$

---

## Forma geral

$$
f(x) = \log_b(x)
$$

onde:

- `b` é a base do logaritmo;
- `x` é o valor de entrada;
- `f(x)` é o resultado.

## 2. Intuição prática

Funções logarítmicas aparecem quando os valores crescem muito rápido e precisamos “comprimir” a escala para enxergar melhor os dados.

### Exemplos do dia a dia:
- escala Richter de terremotos;
- intensidade sonora em decibéis;
- crescimento de dados em tecnologia;
- comparação entre valores muito diferentes.

### Ideia importante:
O logaritmo cresce, mas cresce **devagar**.

Isso é o oposto da função exponencial, que pode crescer muito rapidamente.

## 3. Exemplo básico

Vamos calcular alguns valores da função:

$$
f(x) = \log_2(x)
$$

Lembrando que só podemos calcular logaritmo de valores **positivos**.

```python
import numpy as np

valores = [1, 2, 4, 8, 16, 32]
[(x, np.log2(x)) for x in valores]
```

## Resolução comentada do exemplo

- $$log_2(1) = 0$$, porque \(2^0 = 1\)
\
- $$log_2(2) = 1$$, porque \(2^1 = 2\)
  
- $$log_2(4) = 2$$, porque \(2^2 = 4\)
  
- $$log_2(8) = 3$$, porque \(2^3 = 8\)

- $$log_2(16) = 4$$, porque \(2^4 = 16\)  

\
Perceba que, mesmo dobrando muito o valor de entrada, a saída cresce lentamente.

## 4. Relação entre exponencial e logaritmo

Vamos comparar:

### Exponencial
$$
2^x
$$

### Logarítmica
$$
\log_2(x)
$$

Uma cresce muito rápido.  
A outra cresce devagar.

Isso é essencial para entender dados, algoritmos e IA.

```python
import matplotlib.pyplot as plt

x_exp = np.linspace(0, 6, 200)
y_exp = 2 ** x_exp

x_log = np.linspace(0.1, 20, 400)
y_log = np.log2(x_log)

plt.figure(figsize=(8, 5))
plt.plot(x_exp, y_exp, label="2^x")
plt.plot(x_log, y_log, label="log2(x)")
plt.title("Comparação entre função exponencial e logarítmica")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Valores para a função exponencial
x_exp = np.linspace(-3, 3, 400)
y_exp = 2 ** x_exp

# Valores para a função logarítmica
x_log = np.linspace(0.1, 8, 400)
y_log = np.log2(x_log)

# Valores para a reta y = x
x_reta = np.linspace(-3, 8, 400)
y_reta = x_reta

# Criando o gráfico
plt.figure(figsize=(8, 6))

plt.plot(x_exp, y_exp, label='y = 2^x')
plt.plot(x_log, y_log, label='y = log2(x)')
plt.plot(x_reta, y_reta, '--', label='y = x')

# Eixos
plt.axhline(0)
plt.axvline(0)

plt.xlim(-3, 8)
plt.ylim(-4, 8)

plt.title('Simetria entre a função exponencial e logarítmica')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

plt.show()
```

## O que observar no gráfico?

2. Qual curva cresce mais devagar?
3. Por que o gráfico do logaritmo não começa em zero?
4. O que acontece com valores muito pequenos de `x`?

## 5. Aplicação prática: quantidade de bits para representar valores

Uma aplicação clássica do logaritmo em computação é descobrir quantos bits são necessários para representar uma certa quantidade de valores.

Exemplo:

Se você quer representar 8 valores diferentes, precisa de:

$$
\log_2(8) = 3
$$

Ou seja, **3 bits**.

Vamos testar outros casos.

```python
valores = [2, 4, 8, 16, 32, 64, 128]
[(v, int(np.log2(v))) for v in valores]
```



> “O logaritmo aparece quando queremos descobrir quantas divisões, etapas ou bits são necessários para alcançar certo valor.”

## 7. Mostrando um exemplo ligado à IA e análise de dados

Às vezes temos dados com valores muito distantes entre si.

Exemplo:
- 10
- 100
- 1000
- 10000

Em vez de olhar esses valores na escala normal, podemos aplicar logaritmo para enxergar melhor as diferenças.

```python
dados = np.array([10, 100, 1000, 10000])
log_dados = np.log10(dados)

print("Dados originais:", dados)
print("Dados em escala log10:", log_dados)
```

```python
x = np.arange(len(dados))

plt.figure(figsize=(8, 5))
plt.plot(x, dados, marker="o", label="Escala normal")
plt.plot(x, log_dados, marker="o", label="Escala log10")
plt.title("Comparação entre escala normal e logarítmica")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.legend()
plt.grid()
plt.show()
```

```python
x = np.arange(len(dados))

plt.figure(figsize=(8, 5))
#plt.plot(x, dados, marker="o", label="Escala normal")
plt.plot(x, log_dados, marker="o", label="Escala log10")
plt.title("Comparação entre escala normal e logarítmica")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.legend()
plt.grid()
plt.show()
```

## explicacao



> “Quando os números ficam grandes demais ou muito distantes entre si, o logaritmo ajuda a trazer esses valores para uma escala mais fácil de analisar.”

Isso é muito útil em ciência de dados e IA, porque datasets reais podem ter grande variação de valores.

## 8. Atividade prática com Python

Agora vamos trabalhar com um exemplo inspirado em algoritmos.

Imagine um processo em que, a cada etapa, a quantidade de possibilidades é reduzida pela metade.

Esse tipo de lógica aparece em busca binária e em ideias de eficiência computacional.

Queremos estimar quantas etapas são necessárias para reduzir um problema de tamanho `n`.

```python
# Exemplo 2: número aproximado de etapas para reduzir pela metade

valores_n = [2, 4, 8, 16, 32, 64, 128, 256]
etapas = [np.log2(n) for n in valores_n]

list(zip(valores_n, etapas))
```

```python
plt.figure(figsize=(8, 5))
plt.plot(valores_n, etapas, marker="o")
plt.title("Etapas aproximadas em função de n")
plt.xlabel("Tamanho do problema (n)")
plt.ylabel("log2(n)")
plt.grid()
plt.show()
```
