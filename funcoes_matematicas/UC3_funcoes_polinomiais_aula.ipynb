## 1. Revisão rápida: o que é uma função?

Uma função é uma regra que relaciona uma **entrada** `x` a uma **saída** `f(x)`.

Exemplo:

\[
f(x) = 2x + 3
\]

Quando a expressão possui potências de `x`, como `x²`, `x³` etc., temos uma **função polinomial**.

---

## 2. O que é uma função polinomial?

Uma função polinomial é uma função escrita como soma de termos do tipo:

\[
a_nx^n + a_{n-1}x^{n-1} + \dots + a_1x + a_0
\]

Exemplos:

- \(f(x) = 3x + 1\) → polinomial do 1º grau  
- \(f(x) = x^2 + 2x + 1\) → polinomial do 2º grau  
- \(f(x) = x^3 - 4x\) → polinomial do 3º grau

Nesta aula, vamos focar principalmente na função polinomial do **2º grau**, também chamada de **função quadrática**:

\[
f(x) = ax^2 + bx + c
\]

onde:

- `a` controla a concavidade;
- `b` influencia a inclinação e a posição do vértice;
- `c` é o valor onde a curva corta o eixo `y`.

## 3. Exemplo do cotidiano

Imagine a trajetória de uma bola lançada para cima.  
Ela sobe, atinge uma altura máxima e depois desce.

Esse comportamento não é linear.  
Ele pode ser modelado por uma função quadrática.

Exemplo:

\[
h(t) = -5t^2 + 20t + 1
\]

onde:

- `t` = tempo
- `h(t)` = altura da bola

Perceba:

- o termo `-5t²` faz a curva "abrir para baixo";
- existe um ponto máximo;
- o gráfico não é uma reta, e sim uma **parábola**.

```python
# Exemplo 1: calculando a altura da bola em diferentes instantes

def h(t):
    return -5*t**2 + 20*t + 1

tempos = [0, 1, 2, 3, 4]
[(t, h(t)) for t in tempos]
```

## 4. Representação gráfica

A função quadrática gera, em geral, uma **parábola**.

Vamos visualizar o gráfico do exemplo da bola.

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 4.5, 200)
altura = -5*t**2 + 20*t + 1

plt.figure(figsize=(8, 5))
plt.plot(t, altura)

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

plt.title("Trajetória da bola")
plt.xlabel("Tempo")
plt.ylabel("Altura")
plt.grid()
plt.show()
```

## 5. Entendendo o papel dos coeficientes

Na função:

\[
f(x) = ax^2 + bx + c
\]

### Papel de `a`
- Se `a > 0`, a parábola abre para cima.
- Se `a < 0`, a parábola abre para baixo.

### Papel de `c`
- É o valor da função quando `x = 0`.
- Indica o ponto em que a curva cruza o eixo `y`.

### Papel de `b`
- Ajuda a determinar a posição do vértice e o comportamento da curva.

```python
# Comparando diferentes valores de a

x = np.linspace(-5, 5, 300)

y1 = x**2
y2 = 2*x**2
y3 = -x**2

plt.figure(figsize=(8, 5))
plt.plot(x, y1, label="f(x) = x²")
plt.plot(x, y2, label="f(x) = 2x²")
plt.plot(x, y3, label="f(x) = -x²")
plt.title("Influência do coeficiente a")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid()
plt.show()
```

## 6. Construindo a função a partir de uma situação

### Exemplo 2

Uma empresa está analisando o desempenho de um pequeno experimento de treinamento.  
Ao longo do tempo, o desempenho do modelo melhora no início, atinge um ponto ideal e depois piora por excesso de ajuste em um parâmetro.

Um modelo simplificado para representar esse comportamento é:

\[
p(x) = -x^2 + 6x + 2
\]

onde:

- `x` = valor de ajuste do parâmetro
- `p(x)` = desempenho do modelo

A ideia aqui não é ensinar o treinamento real de IA ainda, mas mostrar que **nem todo comportamento é uma reta**.

```python
# Exemplo 2: avaliando desempenho em diferentes valores do parâmetro

def desempenho(x):
    return -x**2 + 6*x + 2

parametros = [0, 1, 2, 3, 4, 5, 6]
[(x, desempenho(x)) for x in parametros]
```

```python
# Gráfico do desempenho

x = np.linspace(0, 6, 200)
y = -x**2 + 6*x + 2

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title("Desempenho do modelo em função de um parâmetro")
plt.xlabel("Parâmetro")
plt.ylabel("Desempenho")
plt.grid()
plt.show()
```

## 7. Onde isso aparece em IA?

Aqui está a conexão importante desta aula com Inteligência Artificial:

### 1. Relações não lineares nos dados
Nem todo fenômeno do mundo real cresce em linha reta.  
Muitas relações são curvas.

Exemplos:
- evolução de erro e acerto em experimentos;
- impacto de um parâmetro no desempenho do modelo;
- trajetórias, movimentos e sinais;
- ajuste de curvas em análise de dados.

### 2. Engenharia de atributos
No futuro, vocês verão que às vezes usamos variáveis como:

- `x`
- `x²`
- `x³`

para ajudar um modelo de IA a capturar padrões mais complexos.

### 3. Regressão polinomial
Mais adiante, vocês verão uma técnica em que o modelo aprende padrões curvos em vez de apenas uma reta.

**Resumo para guardar:**

> Função polinomial ajuda a modelar comportamentos curvos.  
> Em IA, isso é útil quando os dados não seguem uma relação linear simples.

## 8. Aplicação prática com Python

Agora vamos criar e visualizar outra função polinomial.

```python
# Exemplo 3: custo de erro em função de uma escolha inadequada de parâmetro

def custo(x):
    return x**2 - 4*x + 5

for x in [0, 1, 2, 3, 4]:
    print(f"x = {x} -> custo = {custo(x)}")
```

```python
# Gráfico do custo

x = np.linspace(-1, 5, 300)
y = x**2 - 4*x + 5

plt.figure(figsize=(8, 5))
plt.plot(x, y)
plt.title("Função polinomial: custo em função de x")
plt.xlabel("x")
plt.ylabel("custo(x)")
plt.grid()
plt.show()
```

## 9. Atividade prática

Crie uma função polinomial para representar uma das situações abaixo:

1. a altura de um objeto lançado;
2. o desempenho de um experimento que melhora e depois piora;
3. um custo que diminui até certo ponto e depois volta a subir.

Depois:

- escreva a função;
- calcule alguns valores;
- faça o gráfico no Python;
- explique o que o gráfico representa.

```python

# Crie aqui sua própria função polinomial

def lançamento_objeto(altura_inicial, velocidade_inicial, aceleracao_gravidade, tempo):
    resultado = (altura_inicial + velocidade_inicial*tempo - 0.5*aceleracao_gravidade*tempo**2)
    return resultado if resultado >= 0 else 0


teste = np.arange(0, 10)
[(x, lançamento_objeto(10, 5, 10, x)) for x in teste]
```

```python
import numpy as np
import matplotlib.pyplot as plt

tempos = np.linspace(0, 10, 100)
y = [lançamento_objeto(altura_inicial=10, velocidade_inicial=10, aceleracao_gravidade=10, tempo=t) for t in tempos]

plt.figure(figsize=(8, 5))
plt.plot(tempos, y)
idx = y.index(max(y))
plt.plot(tempos[idx] ,max(y), 'ro')
plt.text(tempos[idx] ,max(y), f'Altura máxima = {max(y):.2f} m²', ha="left")
plt.grid()
plt.show()
```

#A cereja do bolo

Vamos verificar ainda a diferença de graus

```python
import numpy as np
import matplotlib.pyplot as plt

# dados (você pode mudar esses valores)
x = np.array([1, 2, 3, 4, 5])
y = np.array([3, 7, 13, 21, 31])  # comportamento polinomial

# ajuste polinomial grau 2
coeficientes = np.polyfit(x, y, 2)

# cria função a partir dos coeficientes
funcao = np.poly1d(coeficientes)

print("Função encontrada pela IA:")
print(funcao)
```

```python
# gerar pontos para o gráfico
x_linha = np.linspace(1, 5, 100)
y_linha = funcao(x_linha)

plt.scatter(x, y, label="Dados reais")
plt.plot(x_linha, y_linha, color='red', label="Função encontrada")
plt.legend()
plt.grid()
plt.show()
```
